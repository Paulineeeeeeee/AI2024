```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 假設您的數據文件名為 'training_data.txt'
file_path = 'submissions/training_log.txt'  # 請替換為您的文件路徑
# 讀取數據
df = pd.read_csv(file_path, sep=':|,',header=None,names=['0','Step','2','AvgScore','4','ValueLoss'], engine='python')
# 清理和轉換數據
df.columns = [col.strip() for col in df.columns]  # 去除列名中的空白
for col in df.columns:
    if df[col].dtype == object:
        df.drop(col, axis=1, inplace=True)

# 確認數據是否正確轉換
print(df.head())
```

```python
import matplotlib.pyplot as plt

# 假設 df 是您的 DataFrame
# 確保 DataFrame 已經載入，並且 AvgScore 和 ValueLoss 列已準備好

plt.figure(figsize=(10, 5))

# 使用 DataFrame 的索引作為 x 軸
plt.plot(df.index, df['AvgScore'], marker='o', linestyle='-', label='Average Score')

plt.title('Training Progress - Average Score Based on Index')
plt.xlabel('Step')
plt.ylabel('Average Score')
plt.grid(True)
plt.legend()

plt.show()
```
